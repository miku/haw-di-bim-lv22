```python
def clean(token):
    token = token.lower()
    match = re.findall(r'\w+', token)        
    if len(match) > 0:
        return match[0]
```

```python
corpus = collections.defaultdict(list)

for id, name in df[["ID", "Transkript"]].values:
    if not os.path.exists(name):
        continue
    with open(name) as f:
        for line in f:
            for word in line.strip().split():
                corpus[id].append(clean(word))
```

```python
def wordfreq(corpus):
    counter = collections.Counter()
    for token in corpus:
        counter[token] += 1
    return counter
```

```python
for id, tokens in corpus.items():
    num_distinct_tokens = len(wordfreq(tokens))
    ratio = num_distinct_tokens / len(tokens)
    print(id, len(tokens), num_distinct_tokens, ratio)
```

```python
stop_words = set(functools.reduce(set.intersection, [set(corpus[k]) for k in corpus.keys()]))
```
